```python
import pandas as pd
import matplotlib.pyplot as plt
from xgboost import plot_importance


from main import load_data, split_data, train_model, evaluate_model
```

```python
data = load_data('data/dataset.csv')

X_train, X_test, y_train, y_test = split_data(data, 'target70', 0.2)

model = train_model(X_train, y_train, 'target70', 10)

recall, f1, accuracy, precision = evaluate_model(model, X_test, y_test)
```

```python
recall, f1, accuracy, precision
```

```python
importance = model.feature_importances_
```

```python
# 한글깨짐 방지
plt.rc("font", family = "Malgun Gothic")

plot_importance(model, importance_type='weight', title='Weight Importance')
plt.show()

plot_importance(model, importance_type='gain', title='Gain Importance')
plt.show()

plot_importance(model, importance_type='cover', title='Cover Importance')
plt.show()
```
